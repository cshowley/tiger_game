```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tiger_game(games, round_count, count_val, accuracy):
    winnings = [0]
    for i in range(games):
        count = 0
        rounds = 0
        while rounds < round_count:
            while count < count_val:
                winnings[-1] -= 1
                if np.random.random() < (1-accuracy):
                    count -= 1
                else:
                    count += 1
            rounds += 1
            count = 0
        if np.random.random() > (1-(1-accuracy)**count_val):
            winnings.append(winnings[-1]-100)
        else:
            winnings.append(winnings[-1]+10)
    return winnings
```

```python
def plot_results(round_count, accuracy):
    plt.figure()
    plt.hist([tiger_game(1000, round_count, 1, accuracy)[-1] for _ in range(1000)], 50, color='k', edgecolor='black', alpha=0.2, label='surety=1')
    plt.hist([tiger_game(1000, round_count, 2, accuracy)[-1] for _ in range(1000)], 50, color='r', edgecolor='black', alpha=0.2, label='surety=2')
    plt.hist([tiger_game(1000, round_count, 3, accuracy)[-1] for _ in range(1000)], 50, color='b', edgecolor='black', alpha=0.2, label='surety=3')
    plt.hist([tiger_game(1000, round_count, 4, accuracy)[-1] for _ in range(1000)], 50, color='y', edgecolor='black', alpha=0.2, label='surety=4')
    plt.legend(loc='best')
    plt.title('Winnings distribution for %s rounds at %s%% auditory accuracy' % (round_count, int(accuracy*100)))
    plt.xlabel('Winnings ($)')
```

```python
plot_results(1, 0.85)
plot_results(2, 0.85)
plot_results(3, 0.85)
plt.show()
```
